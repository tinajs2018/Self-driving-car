```python
import math


class AutonomousVehicle:
    def __init__(self, position, velocity):
        self.position = position
        self.velocity = velocity

    def update(self, target, obstacle_positions):
        desired_force = self.calculate_desired_force(target)
        obstacle_avoidance_force = self.calculate_obstacle_avoidance_force(obstacle_positions)

        total_force = desired_force + obstacle_avoidance_force

        acceleration = total_force
        self.velocity += acceleration
        self.position += self.velocity

    def calculate_desired_force(self, target):
        desired_velocity = (target - self.position)  # Set desired velocity towards the target
        desired_force = (desired_velocity - self.velocity)  # Set desired force as the difference in velocity
        return desired_force

    def calculate_obstacle_avoidance_force(self, obstacle_positions):
        obstacle_avoidance_force = 0.0

        for obstacle_position in obstacle_positions:
            distance = obstacle_position - self.position
            if distance < 10:  # Apply obstacle avoidance force if obstacle is within a certain range
                repulsive_force = 1 / distance  # Set repulsive force inversely proportional to distance
                obstacle_avoidance_force += repulsive_force

        return obstacle_avoidance_force


# Test scenario
vehicle = AutonomousVehicle(position=0, velocity=0)
target_position = 100
obstacle_positions = [30, 50, 80]

# Perform navigation for a certain number of time steps
for _ in range(100):
    vehicle.update(target_position, obstacle_positions)
    print("Vehicle Position:", vehicle.position)
```
